```python
graph = {"A": ["B"], "B": ["C", "D"], "C": ["D"]}

def find_shortest_path(graph, start, end, path=[]):
    path = path + [start]
    if start == end:
        return path
    shortest = None
    for node in graph[start]:
        if node not in path:
            newpath = find_shortest_path(graph, node, end, path)
            if newpath:
                if not shortest or len(newpath) < len(shortest):
                    shortest = newpath
    return shortest
```

```python
print(find_shortest_path(graph, "A", "D"))
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pylab

G = nx.DiGraph()

# Define Vertices
G.add_node('A')
G.add_node('B')
G.add_node('C')
G.add_node('D')

# Define Edges
G.add_edge('A', 'B')
G.add_edge('A', 'C')
G.add_edge('C', 'A')
G.add_edge('B', 'C')
G.add_edge('D', 'C')

# Phylo.draw_graphviz(G)
nx.draw(G, with_labels=True)
plt.show()
```
